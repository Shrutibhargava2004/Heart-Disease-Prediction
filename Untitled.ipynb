### splitting the dataset int train and test

```python
#standardization of data 
x = df1.drop(columns=['TenYearCHD'])
y = df1['TenYearCHD']
```

```python

x = preprocessing.StandardScaler().fit(x).transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 4)

print ('Train set:', x_train.shape,  y_train.shape)
print ('Test set:', x_test.shape,  y_test.shape)
```

### Exploratory data analysis

```python
plt.figure(figsize=(15,12))
numerical_features = ['age', 'cigsPerDay', 'BPMeds', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']
for i , feature in enumerate(numerical_features):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(x=df1[feature] , y=df1['TenYearCHD'])
    plt.title(f'{feature} vs TenYearCHD')
    plt.xlabel(feature)
    plt.ylabel('TenYearCHD')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='TenYearCHD', data=df1)
plt.show()
```

### Fitting Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(x_train, y_train)
```

```python
pred_y = lg.predict(x_test)
```

### Evaluating

```python
from sklearn.metrics import accuracy_score
print("accuracy of the model is = ", accuracy_score(y_test, pred_y))
```

### Confusion metrics

```python
# Confusion matrix 
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, pred_y)
conf_matrix = pd.DataFrame(data = cm, 
                           columns = ['Predicted:0', 'Predicted:1'], 
                           index =['Actual:0', 'Actual:1'])

plt.figure(figsize = (8, 5))
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens")

plt.show()
print('The details for confusion matrix is =')
print (classification_report(y_test, pred_y))
```

## Model - 2
### training the model by droping the missing values

```python
df2 = df.copy()

df2['cigsPerDay'] = df2['cigsPerDay'].fillna(df2['cigsPerDay'].mean())
df2['BPMeds'] = df2['BPMeds'].fillna(df2['BPMeds'].median())  # Since it's binary, median might be better
df2['totChol'] = df2['totChol'].fillna(df2['totChol'].mean())
df2['BMI'] = df2['BMI'].fillna(df2['BMI'].mean())
df2['heartRate'] = df2['heartRate'].fillna(df2['heartRate'].mean())
df2['glucose'] = df2['glucose'].fillna(df2['glucose'].mean())
```

```python
df2.info()
```

### splitting the dataset int train and test

```python
#standardization of data 
x1 = df2.drop(columns=['TenYearCHD'])
y1 = df2['TenYearCHD']
x1 = preprocessing.StandardScaler().fit(x1).transform(x1)

x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size = 0.3, random_state = 4)

print ('Train set:', x1_train.shape,  y1_train.shape)
print ('Test set:', x1_test.shape,  y1_test.shape)
```

### Fitting Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lg1 = LogisticRegression()
lg1.fit(x1_train, y1_train)
```

```python
pred_y1 = lg1.predict(x1_test)
```

### Evaluating

```python
from sklearn.metrics import accuracy_score
print("accuracy of the model is = ", accuracy_score(y1_test, pred_y1))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Split the dataset
X = df1.drop('TenYearCHD', axis=1)  # Features
y = df1['TenYearCHD']  # Target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize Gaussian Naive Bayes model
gnb = GaussianNB()

# Fit the model
gnb.fit(X_train, y_train)

# Predict on test data
y_pred = gnb.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Confusion matrix and classification report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
